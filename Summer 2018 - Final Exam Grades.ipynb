# Final Exam Grades

Let's use pandas and Matplotlib to explore the final exam grades.

```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('su2018_final_grades.csv')
```

## Histogram of Total Points

```python
hist = df['Total'].hist(bins=10)
#plt.xlim(0, df['Total'].max() + 1)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel("Total Points", fontsize=14)
plt.ylabel("Number of Exams", fontsize=14)
plt.title("Final Grade Distribution", fontsize=15);
```

## Summary Statistics

```python
summary_stats = df['Total'].describe()[1:]
for i, x in zip(summary_stats.index, summary_stats):
    print(i, round(x, 1))
```

## Histograms of Points per Question

```python
hist = df.iloc[:, :11].hist(figsize=(12, 12), bins=10, range=[0, 10])
plt.suptitle("Histograms of Points per Question", fontsize=15);
```

## Mean Points per Question

```python
q_means = df.mean()[:11]

ax = q_means.plot(figsize=(8, 5), grid=True, color='r')

for i, x in enumerate(q_means):
    ax.text(i - .25, x + .5, str(round(x, 1)), color='b', fontsize=12)
ax.lines[-1].set_linewidth(2)
ax.set_xticklabels(df.columns[:11].tolist())

plt.ylim(-1, 11)
plt.xlim(-1, 11)
plt.xticks(range(11), rotation='horizontal', fontsize=12)
plt.yticks(range(11), fontsize=12)
plt.xlabel("Question", fontsize=14)
plt.ylabel("Points", fontsize=14)
plt.title("Mean Points per Question", fontsize=15);
```
