# Midterm Exam Grades

Let's use pandas and Matplotlib to explore the midterm exam grades.

```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('su2018_midterm_grades.csv')
```

## Histogram of Overall Grades

```python
hist = df['Total'].hist(bins=10)
plt.xlabel("Grade")
plt.ylabel("Number of Exams")
plt.title("Midterm Grade Distribution");
```

## Summary Statistics

```python
summary_stats = df['Total'].describe()[1:]
for i, x in zip(summary_stats.index, summary_stats):
    print(i, round(x, 1))
```

## Histogram of Points per Question

```python
hist = df.iloc[:, :11].hist(figsize=(12, 12), bins=10, range=[0, 10])
```

## Mean scores for each question

```python
q_means = df.mean()[:11]

ax = q_means.plot(grid=True, color='r', figsize=(8, 5))

for i, x in enumerate(q_means):
   ax.text(i - .25, x + .5, str(round(x, 1)), color='b', fontsize=12)
ax.lines[-1].set_linewidth(2)

ax.set_ylim(0, 11)
ax.set_xlim(-1, 11)
ax.set_xticks(range(11))
ax.set_yticks(range(1, 11))
ax.set_xticklabels(df.columns[:11].tolist())
ax.set_xlabel("Question", fontsize=14)
ax.set_ylabel("Points", fontsize=14)
plt.xticks(rotation='horizontal', fontsize=12)
plt.yticks(fontsize=12)

ax.set_title("Average Points per Question", fontsize=15);
```
